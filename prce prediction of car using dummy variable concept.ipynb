```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df= pd.read_csv("C:\\Users\\AtulKumar\\Downloads\\CAR_DETAILS_FROM_CAR_DEKHO.csv")
new=df.head(5)
new
```

```python
dm=pd.get_dummies(new.name)
dm
```

```python
concatData= pd.concat([new,dm],axis=1)
concatData
```

```python
final_data=concatData.drop(['name','fuel','seller_type','transmission','owner'],axis=1)
final_data
```

```python
reg = linear_model.LinearRegression()
x = final_data.drop('selling_price',axis=1)
x
```

```python
y= final_data.selling_price
y
```

```python
reg.fit(x,y)
```

```python
Hundai_verna=reg.predict([[2020,50000,0,0,1,0,0]])
Hundai_verna
```

```python
Datsun=reg.predict([[2020,50000,1,0,0,0,0]])
Datsun
```

```python
Honda_Amaze=reg.predict([[2020,50000,0,1,0,0,0]])
Honda_Amaze
```
